## Merge all the excel files into single CSV file


```python
files_path = "./DataScienceProjects/Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/*"
```

```python
files = glob.glob(files_path)
dfs = []
```

```python
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)
combined_files = pd.concat(dfs,ignore_index=True)
combined_files.to_csv('combined_files.csv',index=False)
```

##additional column as a Sales

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data ['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head(10)
all_data.head()
```

## What was the best month for sales? How much was earned?

```python
results = all_data.groupby('Month').sum(numeric_only = True)  #all_data.groupby('Month').sum(numeric_only=True)
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Months')
plt.ticklabel_format(style='plain', axis='y')
plt.show()

#December had highest sales above 4m
```

## Get city from Purchase Address

```python
# Extract state abbreviation using regular expression

# Extract city and state from the 'Purchase Address' column
all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1].strip())
all_data['State'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[2].split()[0])

# Combine city and state in a new column
all_data['City(State)'] = all_data['City'] + '(' + all_data['State'] + ')'
all_data.drop(columns=['City', 'State'], inplace=True)
all_data.rename(columns={'City(State)': 'CITY'}, inplace=True)

all_data.head()
```

## What US city had the highest number of sales?


```python
results = all_data.groupby('CITY').sum(numeric_only = True)
results
```

```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(10, 6))
results.plot(kind='barh', y='Sales', color='orange')
plt.xlabel('Quantity Ordered')
plt.ylabel('City')
plt.title('Sales by City')
plt.legend(['Sales'])

# Show plot
plt.show()
```

## What time of the day Customers order the most?

```python
##get time from order date after
##use quantity ordered and time to plot
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date']) #convert to full date and twenty four hour clock
all_data.head() 
```

```python
all_data['Hour']= all_data['Order Date'].dt.hour #add a hour column
all_data['Minute'] = all_data['Order Date'].dt.minute #add a new column as a Minute
all_data.drop(columns=['minute'], inplace=True) 

all_data.head()
```

```python
import matplotlib.pyplot as plt

# Group data by Hour and count the number of orders
hourly_orders = all_data.groupby('Hour').size()

# Plot the line graph
plt.figure(figsize=(10, 6))
hourly_orders.plot(kind='line', marker='o', color='skyblue')
plt.title('Hourly Order Counts')
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.xticks(range(0, 24))  # Set x-axis ticks to show all hours
plt.grid(True)
plt.show()
```
